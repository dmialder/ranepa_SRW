```python
import pandas as pd
import numpy as np
import datetime
```

```python
share_indexes = pd.read_csv("additional_info/share_indexes.csv")

share_indexes.columns = ['TICKERS']
share_indexes.index=[np.array(share_indexes['TICKERS'])]
share_indexes.to_csv('additional_info/share_indexes.csv', index=False)
```

set 'return' to shares 

считывает из файлов информацию об котировках

вычисляет доходность актива за каждый день по формуле: (data_share['CLOSE'][i]-data_share['CLOSE'][i-1])/data_share['CLOSE'][i-1]

записывает доходность в последний столбец файла с информацией об активе

```python
stock_tickers = pd.read_csv("additional_info/share_indexes.csv")
stock_tickers = pd.Series(stock_tickers['TICKERS'])
for item in stock_tickers:
    data_share = pd.read_csv(f"stock_history/{item}_history.csv")
    share_return = np.array([(data_share['CLOSE'][i]-data_share['CLOSE'][i-1])/data_share['CLOSE'][i-1] 
                            for i in range(1, len(data_share['CLOSE']))])
    share_return = np.insert(share_return, 0, 0)

    data_share['return'] = share_return
    data_share.to_csv(f"stock_history/{item}_history.csv", index=False)
```

set 'return' to bench

аналогично предыдущему пункту высчитывает и записывает в файл бенчмарков их ежедневную доходность

```python
bench_tickers = pd.read_csv("additional_info/bench_tickers.csv")
bench_tickers = pd.Series(bench_tickers['0'])
for item in bench_tickers:
    data_bench = pd.read_csv(f"benchmarks/{item}.csv")
    bench_return = np.array([(data_bench['CLOSE'][i]-data_bench['CLOSE'][i-1])/data_bench['CLOSE'][i-1] 
                            for i in range(1, len(data_bench['CLOSE']))])
    bench_return = np.insert(bench_return, 0, 0)

    data_bench['return'] = bench_return
    data_bench.to_csv(f"benchmarks/{item}.csv", index=False)
```

Calculate VARIANCE of market "variance(Rm)", COVARIANCE Ri of Rm and BETA

В данном окне вычисляется коэффициент бета для каждой акции, для этого

1) вариция бенчмарка (в данном случае беру IMOEX) вычисляется исключительно по тем дням (TRADEDATE), в которые есть информация о акции
и бенчмарка одновременно (к примеру, информация бенчмарков в марте прошлого года практически полностью отсутствует, в то время, как
котировки акций были расписаны за целый март, поэтому этот момент так же пришлось учитывать) 

2) формула вариации: (sum(Rm**2))/N - average(Rm)**2, где Rm - доходность IMOEX

3) формула ковариации:

4) beta = covariance(Ri, Rm) / variance(Rm)

```python
data_bench = pd.read_csv(f"benchmarks/IMOEX.csv")
len_data_bench = len(data_bench['TRADEDATE'])

share_indexes = pd.read_csv('additional_info/share_indexes.csv')
beta = []

for curr_ticker in share_indexes['TICKERS']: # пробегаемся по всем тикерам
    data_share = pd.read_csv(f"stock_history/{curr_ticker}_history.csv")
    len_data_share = len(data_share['TRADEDATE'])

    it = len_data_bench - 1 # это номер массива бенчмарка (идем с конца, с 22го февраля 23 года)
    jt = len_data_share - 1 # это номер массива акции (идем с конца, с 22го февраля 23 года)
    sum_square_return_bench = 0 # сумма квадратов доходностей бенчмарка всех дней (для вычисления вариации)
    sum_bench_return = 0 # сумма доходностей бенчмарка, чтобы вычислить среднюю доходность за день (для вычисления вариации и ковариации)
    sum_share_return = 0 # сумма доходностей акции, чтобы вычислить среднюю доходность за день (для вычисления ковариации)
    sum_mult_return = 0 # сумма произведений доходностей бенчмарка и акции всех дней (для вычисления ковариации)

    day_counter = 0
    while(datetime.date.fromisoformat(data_share['TRADEDATE'][1]) <= datetime.date.fromisoformat(data_bench["TRADEDATE"][it])):
        if data_share['TRADEDATE'][jt] == data_bench['TRADEDATE'][it]:
            sum_square_return_bench += data_bench['return'][it]**2
            sum_bench_return += data_bench['return'][it]
            sum_share_return += data_share['return'][jt]
            sum_mult_return += (data_bench['return'][it] * data_share['return'][jt])

            day_counter += 1
            it -= 1
            jt -= 1
            
        elif data_share['TRADEDATE'][jt] >= data_bench['TRADEDATE'][it]:
            jt -= 1
        elif data_share['TRADEDATE'][jt] <= data_bench['TRADEDATE'][it]:
            it -= 1

    expected_value_bench_return = sum_bench_return / day_counter # средняя доходность бенчмарка в день
    sum_square_return_bench /= day_counter # средний квадрат доходности бенчмарка в день
    variance_Rm = sum_square_return_bench - expected_value_bench_return**2 # вариация

    expected_value_share_return = sum_share_return / day_counter # средняя доходность акции в день
    expected_value_mult_return = sum_mult_return / day_counter # среднее произведение доходности бенчмарка и акции за день

    covariance_Ri_Rm = expected_value_mult_return - expected_value_share_return * expected_value_bench_return # ковариация
    print(curr_ticker)

    beta.append(covariance_Ri_Rm / variance_Rm) # бета

share_indexes['BETA'] = beta
share_indexes.to_csv("additional_info/share_indexes.csv", index=False)
```
